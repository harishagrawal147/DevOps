```python
from datetime import date

# Ask the user for two dates
year1, month1, day1 = map(int, input("Enter the first date (YYYY MM DD): ").split())
year2, month2, day2 = map(int, input("Enter the second date (YYYY MM DD): ").split())

# Create date objects
date1 = date(year1, month1, day1)
date2 = date(year2, month2, day2)

# Calculate the difference
delta = date2 - date1

# Display the result
print("Number of days between the two dates:", delta.days)
```
