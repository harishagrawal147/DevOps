```python
def create_histogram(numbers):
    frequency = {}
    for num in numbers:
        frequency[num] = frequency.get(num, 0) + 1

    for num, count in sorted(frequency.items()):
        print(f"{num}: {'*' * count}")

numbers = [4, 2, 2, 8, 3, 3, 3, 4, 4, 5, 8, 8, 8, 8]
create_histogram(numbers)
```
