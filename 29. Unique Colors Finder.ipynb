```python
def unique_colors(color_list_1, color_list_2):
    result = [color for color in color_list_1 if color not in color_list_2]
    return result

color_list_1 = ["Red", "Green", "Blue", "White", "Black"]
color_list_2 = ["Black", "Yellow", "Green"]

print("Colors in color_list_1 but not in color_list_2:")
print(unique_colors(color_list_1, color_list_2))
```
