```python
def is_member(group, value):
    if value in group:
        return True
    else:
        return False

group = [1, 5, 8, 3, 4, 9]
value = int(input("Enter a value to check: "))

if is_member(group, value):
    print(f"{value} is present in the group.")
else:
    print(f"{value} is not present in the group.")
```
