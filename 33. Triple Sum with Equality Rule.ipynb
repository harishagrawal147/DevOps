```python
def sum_three_integers(a, b, c):
    if a == b or b == c or a == c:
        return 0
    else:
        return a + b + c

x = int(input("Enter first integer: "))
y = int(input("Enter second integer: "))
z = int(input("Enter third integer: "))

print(f"The result is: {sum_three_integers(x, y, z)}")
```
